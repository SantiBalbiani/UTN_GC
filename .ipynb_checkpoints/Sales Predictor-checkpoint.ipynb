```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Leer los datos del archivo CSV
df = pd.read_csv('ventas.csv')

# Extraer los datos de semanas, contactos y ventas
semanas = df['Mes'].values.reshape(-1, 1)
contactos = df['Contactos'].values.reshape(-1, 1)
ventas = df['Ventas'].values

ventas_completas = np.full(9, np.nan)  # Crea un array de longitud 9 lleno de np.nan
ventas_completas[:len(ventas)] = ventas  # Reemplaza los primeros valores con los conocidos
ventas = ventas_completas


ventas = df['Ventas'].values

# Filtrar los datos que tienen valores de ventas
X_train = np.hstack((semanas[:3], contactos[:3]))  # Solo hasta la semana 3
y_train = ventas[:3]  # Solo hasta la semana 3

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Ajustar el modelo con los datos
modelo.fit(X_train, y_train)

# Predecir las ventas para las semanas 4 a 9
X_test = np.hstack((semanas[3:], contactos[3:]))
ventas_prediccion = modelo.predict(X_test)

# Añadir las predicciones al DataFrame original
df.loc[3:, 'Ventas'] = ventas_prediccion

# Imprimir las predicciones
for semana, prediccion in zip(df['Mes'][3:], df['Ventas'][3:]):
    print(f"Predicción de ventas para la semana {semana}: {prediccion:.2f}")

# Visualización
#plt.scatter(semanas[:3], ventas[:3], color='blue', label='Datos de ventas')
#plt.scatter(semanas[3:], df['Ventas'][3:], color='green', label='Predicciones')
#plt.plot(semanas, modelo.predict(np.hstack((semanas, contactos))), color='red', label='Modelo de regresión lineal')
#plt.xlabel('Semana')
#plt.ylabel('Ventas')
#plt.title('Predicción de Ventas')
#plt.legend()
#plt.show()

# Visualización
fig, ax1 = plt.subplots()

# Gráfico de ventas
ax1.scatter(semanas[:3], ventas[:3], color='blue', label='Datos de ventas')
ax1.scatter(semanas[3:], df['Ventas'][3:], color='green', label='Predicciones')
ax1.plot(semanas, modelo.predict(np.hstack((semanas, contactos))), color='red', label='Modelo de regresión lineal')
ax1.set_xlabel('Mes')
ax1.set_ylabel('Ventas', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')

# Crear un segundo eje y
ax2 = ax1.twinx()
ax2.plot(semanas, contactos, color='orange', label='Contactos', linestyle='dashed')
ax2.set_ylabel('Contactos', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')
ax2.legend(loc='upper right')

plt.title('Predicción de Ventas y Contactos')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

# Leer los datos del archivo CSV
df = pd.read_csv('ventas.csv')

# Convertir 'Mes' a string si no lo es ya (para asegurar tipo de datos)
df['Mes'] = df['Mes'].astype(str)

# Codificar 'Mes' usando OneHotEncoder
encoder = OneHotEncoder(sparse=False)
meses_encoded = encoder.fit_transform(df[['Mes']])

# Extraer los datos de semanas, contactos y ventas
semanas = meses_encoded
contactos = df['Contactos'].values.reshape(-1, 1)
ventas = df['Ventas'].values

ventas_completas = np.full(len(df), np.nan)  # Crear un array de NaNs del tamaño de los datos
ventas_completas[:len(ventas)] = ventas  # Reemplazar los valores conocidos de ventas
ventas = ventas_completas

# Filtrar los datos que tienen valores de ventas
X_train = np.hstack((semanas[:3], contactos[:3]))  # Solo hasta la semana 3
y_train = ventas[:3]  # Solo hasta la semana 3

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Ajustar el modelo con los datos
modelo.fit(X_train, y_train)

# Predecir las ventas para las semanas 4 a 9
X_test = np.hstack((semanas[3:], contactos[3:]))
ventas_prediccion = modelo.predict(X_test)


ventas_prediccion_redondeadas = np.round(ventas_prediccion, 2)

# Añadir las predicciones redondeadas al DataFrame original
df.loc[3:, 'Ventas'] = ventas_prediccion_redondeadas


# Imprimir las predicciones
for semana, prediccion in zip(df['Mes'][3:], df['Ventas'][3:]):
    print(f"Predicción de ventas para el mes {semana}: {prediccion:.2f}")

# Visualización
fig, ax1 = plt.subplots(figsize=(8, 4))  # Ajustar el tamaño del gráfico

# Gráfico de ventas
ax1.scatter(np.arange(len(semanas[:3])), ventas[:3], color='blue', label='Datos de ventas')  # Cambiado a np.arange para visualizar
ax1.scatter(np.arange(3, len(semanas)), df['Ventas'][3:], color='green', label='Predicciones')  # Cambiado a np.arange para visualizar
ax1.plot(np.arange(len(semanas)), modelo.predict(np.hstack((semanas, contactos))), color='red', label='Modelo de regresión lineal')  # Cambiado a np.arange para visualizar
ax1.set_xlabel('Mes')
ax1.set_ylabel('Ventas', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')

# Ajustar el límite del eje y para Ventas
min_ventas = min(np.min(ventas[:3]), np.min(df['Ventas'][3:]))
max_ventas = max(np.max(ventas[:3]), np.max(df['Ventas'][3:]))
ax1.set_ylim(min_ventas, max_ventas)

# Crear un segundo eje y
ax2 = ax1.twinx()
ax2.plot(np.arange(len(semanas)), contactos, color='orange', label='Contactos', linestyle='dashed')  # Cambiado a np.arange para visualizar
ax2.set_ylabel('Contactos', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Ajustar el límite del eje y para Contactos (usar los mismos límites que Ventas)
min_contactos = np.min(contactos)
max_contactos = np.max(contactos)
ax2.set_ylim(min_contactos, max_contactos)

# Ajustar el título y mostrar el gráfico
plt.title('Predicción de Ventas y Contactos')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

# Leer los datos del archivo CSV
df = pd.read_csv('ventas.csv')

# Convertir 'Mes' a string si no lo es ya (para asegurar tipo de datos)
df['Mes'] = df['Mes'].astype(str)

# Codificar 'Mes' usando OneHotEncoder
encoder = OneHotEncoder(sparse=False)
meses_encoded = encoder.fit_transform(df[['Mes']])

# Preparar los datos incluyendo contactos
X = np.hstack((meses_encoded, df['Contactos'].values.reshape(-1, 1)))
y = df['Ventas'].values

# Filtrar los datos que tienen valores de ventas
X_train = X[:3]  # Solo hasta la semana 3
y_train = y[:3]   # Solo hasta la semana 3

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Ajustar el modelo con los datos
modelo.fit(X_train, y_train)

# Preparar datos de prueba (semanas 4 a 9)
X_test = X[3:]

# Predecir las ventas para las semanas 4 a 9
ventas_prediccion = modelo.predict(X_test)

# Redondear las predicciones de ventas
ventas_prediccion_redondeadas = np.round(ventas_prediccion, 0)

# Añadir las predicciones redondeadas al DataFrame original
df.loc[3:, 'Ventas'] = ventas_prediccion_redondeadas

# Imprimir las predicciones redondeadas
for semana, prediccion in zip(df['Mes'][3:], df['Ventas'][3:]):
    print(f"Predicción de ventas para el mes {semana}: {prediccion:.2f}")

# Visualización
fig, ax1 = plt.subplots(figsize=(8, 4))  # Ajustar el tamaño del gráfico

# Gráfico de ventas
ax1.scatter(np.arange(len(meses_encoded[:3])), y[:3], color='blue', label='Datos de ventas')
ax1.scatter(np.arange(3, len(meses_encoded)), df['Ventas'][3:], color='green', label='Predicciones')
ax1.plot(np.arange(len(meses_encoded)), modelo.predict(X), color='red', label='Modelo de regresión lineal')  # Usar todas las semanas para predecir
ax1.set_xlabel('Mes')
ax1.set_ylabel('Ventas', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')

# Ajustar el límite del eje y para Ventas
min_ventas = min(np.min(y[:3]), np.min(df['Ventas'][3:]))
max_ventas = max(np.max(y[:3]), np.max(df['Ventas'][3:]))
ax1.set_ylim(min_ventas, max_ventas)

# Crear un segundo eje y
ax2 = ax1.twinx()
ax2.plot(np.arange(len(meses_encoded)), df['Contactos'], color='orange', label='Contactos', linestyle='dashed')
ax2.set_ylabel('Contactos', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Ajustar el título y mostrar el gráfico
plt.title('Predicción de Ventas basada en cantidad de Contactos')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import lightgbm as lgb
import matplotlib.pyplot as plt

# Leer los datos del archivo CSV
df = pd.read_csv('ventas.csv')

# Convertir 'Mes' a string si no lo es ya (para asegurar tipo de datos)
#df['Mes'] = df['Mes'].astype(str)

meses = {'enero': 1, 'febrero': 2, 'marzo': 3, 'abril': 4, 'mayo': 5, 'junio': 6,
         'julio': 7, 'agosto': 8, 'septiembre': 9, 'octubre': 10, 'noviembre': 11, 'diciembre': 12}
df['Mes'] = df['Mes'].map(meses)

# Separar los datos en conjuntos de entrenamiento y prueba
train_data = df[df['Ventas'].notnull()]  # Datos para entrenamiento
test_data = df[df['Ventas'].isnull()]    # Datos para predicción

# Preparar los datos para LightGBM
X_train = train_data[['Contactos']]
y_train = train_data['Ventas']
X_test = test_data[['Contactos']]

# Crear el conjunto de datos de LightGBM
lgb_train = lgb.Dataset(X_train, y_train)

# Especificar parámetros del modelo
params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': 'rmse',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0  # Controla la cantidad de mensajes impresos durante el entrenamiento
}

# Entrenar el modelo
num_boost_round = 100
verbose_eval = 10  # Muestra el progreso cada 10 iteraciones
modelo = lgb.train(params,
                   lgb_train,
                   num_boost_round=num_boost_round )

# Predecir para todos los datos (entrenamiento + prueba)
ventas_prediccion = modelo.predict(X_test)

# Crear una serie con las predicciones y el índice correcto
predicciones_serie = pd.Series(ventas_prediccion, index=test_data.index)

# Añadir las predicciones al DataFrame original solo donde 'Ventas' es nulo
df.loc[df['Ventas'].isnull(), 'Ventas'] = predicciones_serie

# Añadir las predicciones al DataFrame original
#df['Ventas_Pred'] = ventas_prediccion

#df.loc[df['Ventas'].isnull(), 'Ventas'] = df['Ventas_Pred']

# Imprimir las predicciones
# for mes, prediccion in zip(df['Mes'], df['Ventas_Pred']):
#    print(f"Predicción de ventas para el mes {mes}: {prediccion:.2f}")

# Visualización
plt.figure(figsize=(8, 4))
plt.scatter(df['Mes'], df['Ventas'], color='blue', label='Datos de ventas')
plt.scatter(df['Mes'], df['Ventas'], color='green', label='Predicciones')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Predicción de Ventas con LightGBM')
plt.legend()
plt.tight_layout()
plt.show()
```
